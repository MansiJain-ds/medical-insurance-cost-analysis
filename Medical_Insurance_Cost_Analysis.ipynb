To analyze medical insurance data and identify key factors affecting insurance charges.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("insurance[1].csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

### Age Distribution
This plot shows the distribution of policyholders by age.

```python
plt.figure(figsize=(6,4))
sns.histplot(df["age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

### Smoking Impact on Charges
This plot compares medical charges between smokers and non-smokers.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="smoker", y="charges", data=df)
plt.title("Charges for Smokers vs Non-Smokers")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="bmi", y="charges", hue="smoker", data=df)
plt.title("BMI vs Charges")
plt.show()
```

## Dataset Overview

The dataset contains information about individuals including age, gender, BMI, number of children, smoking habits, region, and medical insurance charges.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

### Region vs Insurance Charges
This plot compares medical charges across different regions.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="region", y="charges", data=df)
plt.title("Region vs Insurance Charges")
plt.xticks(rotation=45)
plt.show()
```

### Distribution of Insurance Charges
This histogram shows the distribution of medical insurance charges, which appears right-skewed.

```python
plt.figure(figsize=(6,4))
sns.histplot(df["charges"], bins=30, kde=True)
plt.title("Distribution of Insurance Charges")
plt.show()
```

## Key Insights

1. Smokers incur significantly higher medical charges compared to non-smokers.
2. Medical insurance charges increase with age.
3. Individuals with higher BMI tend to have higher medical expenses.
4. Age and BMI show positive correlation with insurance charges.
5. Smoking is the most influential factor affecting medical cost.

## Conclusion

The analysis shows that smoking is the most significant factor influencing medical insurance charges. Age and BMI also positively impact insurance cost. Lifestyle factors play a major role in determining medical expenses.
